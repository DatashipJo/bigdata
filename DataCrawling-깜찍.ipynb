```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
# 페이지 접속
browser = webdriver.Chrome()
url = "https://www.yogiyo.co.kr/mobile/#/"
browser.maximize_window()
browser.get(url)
time.sleep(3)

location = '서울특별시 성동구 행당동 7 성동구청'
elem = browser.find_element_by_name("address_input")
elem.clear()
elem.send_keys(location)
time.sleep(1)
search_xpath = '''//*[@id="button_search_address"]/button[2]'''
search = browser.find_element_by_xpath(search_xpath)
browser.execute_script("arguments[0].click();", search)
time.sleep(2)
```

```python
# 카테고리 선택
# 일단 치킨만!
cg = browser.find_element_by_xpath('/html/body/div[6]/div/div[1]/div[1]/div/div/div/ul/li[5]')
#cg = browser.find_element_by_xpath('/html/body/div[7]/div/div[5]/a')
cg_title = cg.text
cg.click()
```

```python
# 음식점 정렬
time.sleep(2)
browser.find_element_by_xpath('//*[@id="content"]/div/div[1]/div[2]/div/select').click()
time.sleep(1)
browser.find_element_by_xpath('//*[@id="content"]/div/div[1]/div[2]/div/select/option[3]').click()
```

```python
# 치킨집 선택
# i = 1 :: 리뷰 제일 많은 치킨집 선택
i = 1 
browser.find_element_by_xpath('/html/body/div[6]/div/div[5]/div/div/div['+str(i)+']/div/table/tbody/tr/td[2]/div/div[1]').click()
```

```python
# 정보클릭
browser.find_element_by_xpath('//*[@id="content"]/div[2]/div[1]/ul/li[3]/a').click()

# 정보창에 있는 가게 주소를 ADDRESS 변수에 저장
ADDRESS = browser.find_element_by_xpath('//*[@id="content"]/div[2]/div[1]/div[6]/div[2]/p[3]/span').text

# 클린리뷰 클릭
browser.find_element_by_xpath('//*[@id="content"]/div[2]/div[1]/ul/li[2]/a').click()

# 클린리뷰에 있는 총 평점을 TOTAL_STAR 변수에 저장
TOTAL_STAR = browser.find_element_by_xpath('/html/body/div[6]/div[2]/div[1]/div[5]/div[1]/div/div/strong').text

# 가게명을 STORE 변수에 저장
STORE = browser.find_element_by_xpath('/html/body/div[6]/div[2]/div[1]/div[1]/div[1]/span').text
```

```python
# 더보기 클릭
cnt = 0
while True :
    try :
        browser.find_element_by_class_name('btn-more').click()
        time.sleep(1.5)
        cnt += 1
        print("더보기 클릭 {}".format(cnt))
        # 더보기 개수제한
        #if cnt == 10:
            #break
    except:
        print("더보기 종료")
        break
```

```python
# 웹 페이지에서
# 아이디, 날짜, 평점, 리뷰 가져오기

import time

start = time.time()

cnt = 1
while True:
    cnt += 1
    ID = browser.find_elements_by_xpath("/html/body/div[6]/div[2]/div[1]/div[5]/ul/li['+str(cnt)+']/div[1]/span[1]")
    DATE = browser.find_elements_by_xpath("/html/body/div[6]/div[2]/div[1]/div[5]/ul/li['+str(cnt)+']/div[1]/span[2]")
    STAR = browser.find_elements_by_xpath("/html/body/div[6]/div[2]/div[1]/div[5]/ul/li['+str(cnt)+']/div[2]/div/span[2]")
    REVIEW = browser.find_elements_by_xpath("/html/body/div[6]/div[2]/div[1]/div[5]/ul/li['+str(cnt)+']/p")
    del REVIEW[-1]
    
    print('총 ID :',len(ID))
    print('총 DATE :',len(DATE))
    print('총 REVIEW :',len(REVIEW))
    print('총 STAR :',len(STAR))
    break
   
end = time.time()
print(format(end-start))
```

```python
# 정보 가져온걸로 딕셔너리 생성 후 리스트 변수 dict 에 저장
# dict 를 DataFrame으로 변환

start = time.time()

dict = []
for i in range(len(ID)):
    dict.append({
        '아이디' : ID[i].text,
        '날짜' : DATE[i].text,
        '리뷰' : REVIEW[i].text,
        '평점' : STAR[i].text
        
    })

df = pd.DataFrame(dict)
df.head()

end = time.time()
print(format(end-start))
```

```python
# 가게명 추가
df.insert(loc=0, column='가게명', value=STORE)
```

```python
# 가게주소 추가
df.insert(loc=1, column='가게주소', value=ADDRESS)
```

```python
# 총 평점 추가
df.insert(loc=2, column='총평점', value=TOTAL_STAR)
df
```

```python
print(df.head())
# {STORE}로 csv 저장
print(f"{STORE}")

df.to_csv(f'{STORE}.csv',mode='a', header=False, encoding='utf-8-sig')
print("저장완료")
```
